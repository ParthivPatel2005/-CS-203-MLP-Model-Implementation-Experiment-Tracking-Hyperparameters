## Epochs vs Performance 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("results_csv.csv")

# Line plot: Epochs vs. Performance (accuracy and F1)
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x="Epochs", y="Score", hue="Metric", style="Strategy", marker="o", linewidth=2.5)
plt.xlabel("Epochs")
plt.ylabel("Performance (Accuracy / F1)")
plt.title("Effect of Epochs on Accuracy and F1-Score")
plt.legend(title="Metric & Strategy")
plt.grid(True)
plt.show()
```

## Learning Rate vs. Performance (Accuracy & F1-score)
(Note it is difficult to make inference from this graph due to the less number of hyperparameters to optimise, and also the similarity in their predictions)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("results_csv.csv")

df["Learning Rate"] = df["Learning Rate"].astype(str)

# Plot
plt.figure(figsize=(10, 6))
sns.lineplot(
    data=df, 
    x="Learning Rate", 
    y="Score", 
    hue="Metric", 
    style="Strategy", 
    markers=True, 
    dashes=False, 
    linewidth=2.5
)

plt.xlabel("Learning Rate (Log Scale)")
plt.ylabel("Performance (Accuracy / F1)")
plt.title("Effect of Learning Rate on Accuracy and F1-Score")
plt.xscale("log")  # Log-scale for better visualization
plt.legend(title="Metric & Strategy", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.grid(True, linestyle="--", alpha=0.7)
plt.xticks(rotation=45)  
plt.show()
```

## Strategy wise Performance 

```python
plt.figure(figsize=(12, 6))

sns.barplot(
    data=df, 
    x="Strategy", 
    y="Score", 
    hue="Metric", 
    ci=None, 
    palette=sns.color_palette("Set2") 
)

plt.xlabel("Search Strategy", fontsize=12)
plt.ylabel("Score", fontsize=12)
plt.title("Performance Comparison Across Hyperparameter Search Strategies", fontsize=14)

plt.legend(title="Metric", bbox_to_anchor=(1.05, 1), loc="upper left")

plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.xticks(rotation=30, ha="right")

plt.show()
```
